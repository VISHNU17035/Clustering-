```python
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/vishn/Downloads/Customer data_clustering.csv")
df.head()
```

```python
df.tail()
```

```python
# Encode Gender column
df["Gender"] = LabelEncoder().fit_transform(df["Gender"])
```

```python
# Select relevant features for clustering
X_kmeans = df[["Gender", "Age", "Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
# Apply K-Means clustering with optimal k=5
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df["KMeans_Cluster"] = kmeans.fit_predict(X_kmeans)
```

```python
# Calculate Silhouette Score for K-Means
kmeans_silhouette = silhouette_score(X_kmeans, df["KMeans_Cluster"])
print(f"Silhouette Score for K-Means: {kmeans_silhouette:.2f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
 x=df["Annual Income (k$)"],
 y=df["Spending Score (1-100)"],
 hue=df["KMeans_Cluster"],
 palette="Set1",
 alpha=0.8
)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title(f"K-Means Clustering (k={optimal_k}) - Silhouette Score: {kmeans_silhouette:.2f}")
plt.legend(title="Cluster")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler


# Encode categorical variables (if 'Gender' exists)
if 'Gender' in df.columns:
    df['Gender'] = LabelEncoder().fit_transform(df['Gender'])

# Select correct features based on actual column names
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
X = df[features].values  # Convert to NumPy array

# Normalize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Compute Cosine Distance (1 - Cosine Similarity)
cosine_distances = 1 - cosine_similarity(X_scaled)

# Normalize to ensure all values are non-negative
scaler = MinMaxScaler(feature_range=(0, 1))
cosine_distances = scaler.fit_transform(cosine_distances)

# Apply DBSCAN Clustering (Cosine distance)
dbscan = DBSCAN(metric="precomputed", eps=0.3, min_samples=5)
df['DBSCAN_Cluster'] = dbscan.fit_predict(cosine_distances)

# Silhouette Score (if valid clusters are formed)
unique_clusters = set(df['DBSCAN_Cluster'])
if len(unique_clusters) > 1 and -1 in unique_clusters:
    unique_clusters.remove(-1)  # Ignore noise points
if len(unique_clusters) > 1:
    dbscan_silhouette = silhouette_score(X_scaled, df['DBSCAN_Cluster'])
    print(f"DBSCAN Silhouette Score: {dbscan_silhouette:.3f}")
else:
    print("DBSCAN did not form valid clusters.")

# Visualization
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Annual Income (k$)'], y=df['Spending Score (1-100)'], hue=df['DBSCAN_Cluster'], palette="viridis", s=100)
plt.title("DBSCAN Clustering (Cosine Distance)")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend(title="Cluster")
plt.show()

```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Encode categorical variables (if applicable)
if 'Gender' in df.columns:
    df['Gender'] = LabelEncoder().fit_transform(df['Gender'])

# Select correct features
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
X = df[features].values  # Convert to NumPy array

# Normalize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply Agglomerative Clustering (Ward Linkage + Euclidean Distance)
agglo = AgglomerativeClustering(n_clusters=5, linkage="ward", metric="euclidean")
df["Agglomerative_Cluster"] = agglo.fit_predict(X_scaled)

# Calculate Silhouette Score
agglo_silhouette = silhouette_score(X_scaled, df["Agglomerative_Cluster"], metric="euclidean")
print(f"Silhouette Score for Agglomerative Clustering: {agglo_silhouette:.2f}")

# Scatter Plot for Clustering
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df["Annual Income (k$)"],
    y=df["Spending Score (1-100)"],
    hue=df["Agglomerative_Cluster"],
    palette="coolwarm",
    alpha=0.8,
    s=100
)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title(f"Agglomerative Clustering - Silhouette Score: {agglo_silhouette:.2f}")
plt.legend(title="Cluster")
plt.show()

# Dendrogram for Hierarchical Clustering
plt.figure(figsize=(40, 6))
linked = linkage(X_scaled, method='ward')
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Dendrogram for Agglomerative Clustering")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()



```
